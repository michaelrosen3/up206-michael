# Exploring Oakland tree data

Our group is concerned with trees and the inequity of their distribution across the city of Oakland. In this assignment, I'll begin my initial foray into the tree dataset, which is large and unwieldy. By the end, I'll hope to make the data a bit more manageable and perhaps even produce a few maps.

To start, I'll import the libraries necessary for my exploration.

```python
import geopandas as gpd
import pandas as pd
import matplotlib as plt
import contextily as ctx
```

Now time to bring in the data.

```python
oaktrees=gpd.read_file("Trees_Sidewalk_Raw.geojson")
```

Huh, that took a while. I wonder why...

```python
oaktrees.info()
```

Well, that explains it — there are over 42,000 rows and 50 columns! I'm immediately curious about a few of these variables, so I'm going to run a few value_counts commands to see what we're dealing with here.

```python
oaktrees['DEADWOOD'].value_counts()
```

```python
oaktrees['TRUNKDIAM'].value_counts()
```

```python
oaktrees['COUNCILDIS'].value_counts()
```

```python
oaktrees['SPECIES'].value_counts()
```

I'm intrigued by this "SPECIES" variable, and I suspect there are a bunch of values that I'm not seeing right now, so I'll run a little command so all is revealed.

```python
pd.set_option('display.max_rows', None)
```

```python
oaktrees['SPECIES'].value_counts()
```

Oh man, that's a lot of different types of trees. If I'm going to map the different species, I think this will be too many to produce a cogent map from. I'm going to begin trimming (heh) the tree dataset to make it more manageable.

```python
columns=['COUNCILDIS','SPECIES','ADDRNUM','STNAME','geometry']
oaktrees_trimmed=oaktrees[columns]
oaktrees_trimmed.head()
```

Now I've got a new dataframe — oaktrees_trimmed — with just a few key variables. But I still think there are too many trees. Let's see what happens when I try to plot them.

```python
oaktrees_trimmed.plot(figsize=(12,12),
                 column='SPECIES',
                 legend=True, 
                 cmap='coolwarm'
        )
```

Oh no, that's bad news! I need way fewer trees! I'm going to choose a select few species to bring into my map instead. To do so, I'm going to have to use a new method for trimming a dataset. First, I'm going to make a list of the top tree types. Let's bring up the value counts again so I can make my choices...

```python
oaktrees_trimmed.SPECIES.value_counts()
```

I'm going to leave out "tree well only" out of my list because I'm more concerned with actual trees. Otherwise, I'm going to grab the top species.

```python
list=['Platanus acerifolia','Liquidambar styracifula','Pyrus calleryana cvs','Prunus cerasifera/blireiana','Lagerstroemia indica','Pyrus kawakami','Magnolia grandiflora']
series=oaktrees_trimmed.SPECIES.isin(list)
oaktrees_new=oaktrees_trimmed[series]
```

Okay, that should've created a limited dataset with just my desired tree species. Let's run a values_count just to check...

```python
oaktrees_new.SPECIES.value_counts()
```

Beautiful! Let's try our map again.

```python
oaktrees_new.plot(figsize=(12,12),
                 column='SPECIES',
                 legend=True, 
                 cmap='coolwarm'
        )
```

Would you look at that! I can't say that this is a particularly revealing map on its own, but I think it's a nice start to exploring the species data.

But I'm not sure I'm quite done with the species data. I noticed on our latest value_counts run through that there is a single banana tree in Oakland. I want to know where I could find that banana tree. I'm going to create a new "dataframe" with just the banana tree.

```python
banana_list=['Banana']
banana_series=oaktrees_trimmed.SPECIES.isin(banana_list)
banana_tree=oaktrees_trimmed[banana_series]
```

```python
banana_tree.head()
```

There she is — the single banana tree. Let's see if I can map the banana tree with a base map to ascertain its location.

```python
ax = banana_tree.plot(color="yellow", figsize=(9, 9))
ctx.add_basemap(ax, crs=banana_tree.crs.to_string())
```

Hmm... seems like that could conceivably be correct, but I think we might need a bit more zoom.

```python
ax = banana_tree.plot(color="yellow", figsize=(12, 12))
ctx.add_basemap(ax, reset_extent=False,crs=banana_tree.crs.to_string())
```
